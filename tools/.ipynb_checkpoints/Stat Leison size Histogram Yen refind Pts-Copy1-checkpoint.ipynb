```python
import os
import numpy as np

import pandas as pd

folder = '/mnt/fast-data/mjc/AutoRECIST/Inputs/'

df_CTs = pd.read_excel(folder+'AutoRECIST_List_LesionSize_20220602_JM_SingleCTSeries.xlsx')

df_Yen = pd.read_excel(folder+'PDS_AUTO_RECIST CIA-LAB Testing Dataset Gold Standard_Yen_2022-06-13.xlsx')
```

```python
dfinnermerge = pd.merge(df_Yen,df_CTs,how='inner',on='Image File Path')
```

```python
def getValues(df , value='Uni' , condition='Location' , condition_value = ['liver']):
    df_select = df[df[condition].isin(condition_value)]
    return df_select[value].tolist()
import numpy as np
import matplotlib.pyplot as plt
# plt.style.use('seaborn-deep')


bins = np.linspace(0, 100, 11)

x = getValues(dfinnermerge, condition='dataset' , condition_value = ['CUIMCfirst'] )
y = getValues(dfinnermerge, condition='dataset' , condition_value = ['CUIMClast'])
z = getValues(dfinnermerge, condition='dataset' , condition_value = ['Amgen'])

plt.hist([x,y,z], bins, label=[ 'CU_first','CU_last','Amgen_20020408'],color =[(1, 0.3, 0), (0, 0.8, 0), (0, 0, 1)] , lw=0)
plt.legend(loc='upper right')

plt.savefig('ThreeHists_liver_Yen.eps', format='eps')
```
