```python
def check_src_file_exists(masks_w_ap , masks_dir):
    N=0
    for line in masks_w_ap:
        id = line[0]
        src = os.path.join(masks_dir+id , 'mask.npy')
        if not os.path.exists(src):
            print(id,'not exists')
        else:
            N+=1
    print('{} files exist'.format(N))
    
```

```python
from shutil import copyfile,copytree
import os
import glob

evo_sample_path = "../eval_results/seg_superscalenet_lesion_*"
masks_dir = "../output/Lesion/superscalenet_seg/Lesion_Q5_9Slices_scalenet_seg_test/data_patch_test_"

txt_files = glob.glob(evo_sample_path)
txt_files.sort()

def getid(s, l='_f' , r= '.txt'):
    ix1 = s.rfind(l)
    ix2 = s.rfind(r)
    return s[ix1+1:ix2]

masks_w_ap = []
for txt in txt_files:
    with open(txt,'r') as f:
        lines = f.readlines()  
        for lines_ind in range(0, len(lines)):
            line = lines[lines_ind]
            _line = line.split()
            if len(_line) == 15:
                # seed 266555 Params(M) 16.91 FLOPs(G) 229.07  AP 0.7646
                seed = _line[-7]
                id = getid(txt)+'_'+seed
                params = float(_line[-5])
                flops = float(_line[-3])
                ap = float(_line[-1])
                masks_w_ap.append([id, params, flops, ap])
            else:
                print(f"line should be 15 but now is {len(_line)}\n")
```

```python
check_src_file_exists(masks_w_ap , masks_dir)
```

```python
masks_w_ap.sort(key=lambda x: -x[3])
masks_w_ap = masks_w_ap[:1000]
with open('../evo_files/lesion_evo_file.txt' , 'w') as ff:
    for line in masks_w_ap:
        ss = "".join(str(x)+' ' for x in line)
        ff.write(ss+'\n')
```

```python
folder = '../evo_files/masks/'
if not os.path.exists(folder):
    os.mkdir(folder)

for line in masks_w_ap:
    id = line[0]
    src = os.path.join(masks_dir+id , 'mask.npy')
    dst = folder +id+'.npy'
    if os.path.exists(dst):
        print(id,'skip')
        continue
    else:
        print(id,'copying')
    copyfile(src, dst)
```
